```python
import pandas as pd
data=pd.read_csv('ekonomi.csv')

# Rename the columns before grouping
data.rename(columns={'Apakah Anda sudah menerapkan prinsip 9R tersebut? ': 'penerapanprinsip9r'}, inplace=True)


# Calculate the count per gender
rata_rata_per_gender = data.groupby('Gender')['penerapanprinsip9r'].count()
print(rata_rata_per_gender)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='penerapanprinsip9r', data=data)
plt.title('Jumlah Penerapan Prinsip 3R berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Jumlah')
plt.legend(title='Penerapan Prinsip 3R')
plt.show()
```

```python
import pandas as pd

data=pd.read_csv('ekonomi.csv')

# Check if the column exists
if 'penerapan_laranganplastik' not in data.columns:
    # Assuming the column is named 'Apakah Anda sudah mulai menerapkan peraturan tersebut?'
    data.rename(columns={'Apakah Anda sudah mulai menerapkan peraturan tersebut?': 'penerapan_laranganplastik'}, inplace=True)
else:
    print("Column 'penerapan_laranganplastik' already exists.")

# Calculate the count per gender
rata_rata_per_gender = data.groupby('Gender')['penerapan_laranganplastik'].count()
print(rata_rata_per_gender)
```
